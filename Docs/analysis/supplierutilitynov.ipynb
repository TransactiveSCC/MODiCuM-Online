```python
n, i, z = symbols('n, i, z', integer=True, positive=True)
P_d, pi_s, pi_sd, pi_se, pi_a, pi_m, pi_v = symbols('P_d, pi_s, pi_sd, pi_se, pi_a, pi_m, pi_v')
lambd, delta, rho = symbols('lambd, delta, rho', positive=True)
```

```python
US = -P_d*i*pi_s - P_d*pi_sd + i*pi_s -z*pi_se - pi_a - pi_m - pi_v
display(US.expand())
```

```python
total_ways = binomial(i, z)
safe_ways = binomial(i-n, z-n)
P_d_val = 1 - safe_ways/total_ways

pi_sd_val = pi_s*i*rho

US2 = US.subs({P_d: P_d_val, pi_sd: pi_sd_val})
US2
```

```python
result = {}


#z_vals = [299500, 299600, 299700, 299800, 299900, 300000]
per_proc = [0, 0.5, 0.75, 0.88, .94, .97, .99, .995, .996, .997, .998, .999, 1]
z_vals = [int(per_proc[x]*i_val) for x in range(len(per_proc))]
rho_vals = [0.0001, 0.1, .2]
col_names = ['a', 'b', 'c']
q = 0
for rho_val in rho_vals:
    y = []
    for z_val in z_vals:
        US3 = US2.subs({pi_s: pi_s_val,
                       lambd: lambd_val,
                       delta: delta_val,
                       rho: rho_val,
                       i: i_val,
                       n: n_val,
                       z: z_val,
                       pi_a: pi_a_val,
                       pi_m: pi_m_val,
                       pi_se: pi_se_val,
                       pi_v: pi_v_val})
        r = US3.evalf()
        y.append(r)
        total_ways = binomial(i_val, z_val)
        safe_ways = binomial(i_val-n_val, z_val-n_val)
        P_detected = 1 - safe_ways/total_ways
        print(f"P_detected: {P_detected.evalf()}, z_val: {z_val}, n_val: {n_val}, i_val: {i_val}")
    result[col_names[q]] = y
    q += 1
df = pd.DataFrame(result)
df['proc'] = [100* x for x in per_proc]
df['procper'] = [100* x for x in per_proc]
df
```
